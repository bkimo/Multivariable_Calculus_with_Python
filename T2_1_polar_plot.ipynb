```python
# Import necessary packages
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Axis in polar coordinates###

To put the axis in polar coordinates in matplotlib, a solution is to use the option `projection='polar'`:

```python
plt.figure(figsize=(5,5))

plt.subplot(111, projection='polar')

# (Optional) To change the ticks and y-limits of the axis:
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,4)
plt.yticks(np.arange(0,4,1.0))

plt.show()
```

## Plot a point## 

The numpy function `deg2rad()` can be used to transform degrees to radians

```python
plt.figure(figsize=(6,6))

# Plot a point (r, theta) = (2.0, pi/3)
r = 2.0
theta = np.deg2rad(60.0)

plt.subplot(111, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,4)
plt.yticks(np.arange(0,4,1.0))

plt.scatter(theta, r)

#plt.savefig("polar_coordinates_01.png", bbox_inches='tight')

plt.show()
```

### Plot  a function

Plot of the function $r = \theta$ on $[0, 4\pi]$

```python
theta = np.linspace(0, 4*np.pi, 100)
r = theta

plt.figure(figsize=(6,6))

plt.subplot(111, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))

plt.plot(theta, r)

#plt.savefig("polar_coordinates_02.png", bbox_inches='tight')

plt.show()
```

### Polar Plot of a Circle###

The equation of the circle in polar coordinates is given by $r = R$, where $R$ is a radius of the circle.

```python
theta = np.linspace(0, 2*np.pi, 100)
r = 5* np.ones(100)  # radius = 5

plt.figure(figsize=(6,6))

plt.subplot(111, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/2.0))

plt.title('Circle in polar format: $r=R$')

plt.plot(theta, r, 'r')

#plt.savefig("polar_coordinates_02.png", bbox_inches='tight')

plt.show()
```

### Cardioids-Symmetrical around x- and y-axis###

When the radii of two circles are equal, and when one circle is fixed and the other circle is rolling on the first one – a point on the rolling circle traced will plot a cardioid.
 The polar forms of the cardioids are given by
  $$
   \begin{array}{ccc}
    r &=& a + b\cos(k\theta) \\
    r &=& a + b\sin(k\theta)
   \end{array}
 $$
 
When $a=1, b=1$ and $k=1$, the cardioids are symmetrical around $x$- and $y$-axis on the positive side.

```python
# Radian values upto 2*pi
theta = np.linspace(0, (2*np.pi), 100)
a = 1
b = 1
k =1

r1 = a + (b*np.cos(k*theta))
r2 = a + (b*np.sin(k*theta))

plt.figure(figsize=(15,8))

ax1 = plt.subplot(121, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r1,'o')

plt.title('Cardioids in polar format: $r = a + (b*cos(k*theta))$')



ax2 = plt.subplot(122, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r2,'o')

plt.title('Cardioids in polar format: $r = a + (b*sin(k*theta))$')


plt.show()
```

### $n$-leaf Roses###  

Cardioids that are symmetric around y axis. While keeping k > 1, and a=1,b=1,a=2,b=2 k-leaf roses are plotted.

```python
# Radian values upto 2*pi
theta = np.linspace(0, (2*np.pi), 100)
a = 1
b = 1
k1 = 3
k2 = 4

r1 = a + (b*np.cos(k1*theta))
r2 = a + (b*np.cos(k2*theta))

r3 = a + (b*np.sin(k1*theta))
r4 = a + (b*np.sin(k2*theta))

plt.figure(figsize=(15,15))

ax1 = plt.subplot(221, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r1)

plt.title('Cardioids in polar format: $r = a + (b*cos(k*theta))$')



ax2 = plt.subplot(222, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r2)

plt.title('Cardioids in polar format: $r = a + (b*cos(k*theta))$')

ax3 = plt.subplot(223, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r3)

plt.title('Cardioids in polar format: $r = a + (b*sin(k*theta))$')



ax4 = plt.subplot(224, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r4)

plt.title('Cardioids in polar format: $r = a + (b*sin(k*theta))$')



plt.show()
```

**Remark:** In matplotlib, polar plots are based on clipping of the curve. 
For example, in [Fig. 2.7](https://matplotlibguide.readthedocs.io/en/latest/Matplotlib/types.html)
(generated by line 12 in Listing 2.6), only two lobes of 
$cos(2x)$ are generated instead of four. Other two lobes have negative value 
of `r`, therefore these are clipped by the matplotlib. 
The actual four lobes are shown in Fig. 2.8, which can not be generated by matplotlib.

```python
# Radian values upto 2*pi
theta = np.linspace(0, (2*np.pi), 100)
a = 1
b = 1
k =1

r1 = a + (b*np.cos(k*theta))
r2 = a + (b*np.sin(k*theta))

plt.figure(figsize=(15,8))

plt.subplot(111, projection='polar')
plt.xticks(np.arange(0,2.0*np.pi,np.pi/6.0))
plt.ylim(0,2.5)
plt.yticks(np.arange(0,2.5,0.5))

plt.plot(theta,r1)
plt.plot(theta,r2)

plt.title('Two Cardioids on One Window')


plt.show()
```
